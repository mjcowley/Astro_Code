### Lists

Lists are Python's most flexible container. They can hold mixed types, including other lists.


````python
my_list = [1, "a_string", True, 3.14, [2, 4, 6]]
print(my_list)
````

The above list is somewhat chaotic. It's rare to want or need a list with such diverse contents. However, it serves to demonstrate that Python is flexible about the types of items you can include in a list. Beyond manually specifying list contents, Python offers functions that can automatically generate lists, especially useful for creating sequences with a regular pattern:

````python
# Generate a list using a list comprehension
even_numbers = [x for x in range(2, 21, 2)]
print(even_numbers)
````

This Python snippet demonstrates the use of a list comprehension to generate a list of even numbers. Here's how it works:

- `range(2, 21, 2)` creates a sequence of numbers starting from 2 up to (but not including) 21, with a step of 2. This step ensures that only even numbers are included.
- `[x for x in range(2, 21, 2)]` iterates over each number `x` in the sequence, adding `x` to the list.
- The result is a list of even numbers from 2 to 20, which is then printed out.

Let's inspect the lists further:

````python
# Summation and length of the list
list_sum = sum(even_numbers)
print(list_sum)
list_length = len(even_numbers)
print(list_length)
````

List comprehensions offer a concise way to create lists, making this method both efficient and easy to read.

#### Indexing

List elements are accessed by index (zero-based). Negative indices count from the end.


````python
print(my_list[0])       # first element
print(my_list[0:2])     # slice: first two elements
print(my_list[-1])      # last element
print(my_list[-1][1])   # second element of the nested list
````

So, to summarise, indexing allows you to access individual elements of an iterable, such as a list or a string, using their position.

- **Positive Indexing**: Starts from 0 at the beginning of the iterable and increases by 1 for each subsequent element.
- **Negative Indexing**: Starts from -1 for the last element, -2 for the second last, and so on, making it easy to access elements from the end.

### Data Cleaning

Raw data typically requires cleaning before analysis. The three most common tasks are handling missing values, removing duplicates, and standardising formats.

### Handling Missing Values

`isnull().sum()` tells you how many missing values are in each column. A heatmap is useful for spotting where they cluster.


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming co_df is loaded from the exercise above
missing_values_count = co_df.isnull().sum()
print("Missing values per column:\n", missing_values_count)

plt.figure(figsize=(10, 4))
sns.heatmap(co_df.isnull(), cbar=False, cmap='Reds')
plt.title('Missing Value Map')
plt.show()
````

Two common strategies for dealing with missing values:

**1. Imputation**: replace missing values with a summary statistic (mean, median) or a domain-informed value.
**2. Removal**: drop rows with missing values entirely.

Imputation preserves sample size; removal is safer when missing data is not random. Choose based on context.


````python
import numpy as np

# Imputation with column means
numeric_means = co_df.select_dtypes(include=[np.number]).mean()
co_df_imputed = co_df.fillna(numeric_means)
print(co_df_imputed.isnull().sum())
````

````python
# Alternatively: remove rows with any missing value
co_df_dropped = co_df.dropna()
print(co_df_dropped.info())
````

### Removing Duplicates


Let's walk through a quick exercise to remove duplicates. Let's reload our DataFrame and inspect if there are any:

````python
# Reload the original dataset so we're starting fresh with missing values intact
co_df = pd.read_csv('carbonmonoxide-qld-2022.csv', sep=',')
# Report how many rows we have before cleaning
print("Rows:", len(co_df))
# Count rows that are duplicates based on the Date and Time columns combined;
print("Duplicate count:", co_df.duplicated(subset=['Date', 'Time']).sum())
````

Well...there are none. Let's manipulate the data to add some!

````python
# Select the first two rows to use as our artificial duplicates
duplicates_to_add = co_df.iloc[0:2]

# Append those two rows to the end of the DataFrame
co_df_with_dupes = pd.concat([co_df, duplicates_to_add], ignore_index=True)

# Shuffle the DataFrame so the duplicates aren't obviously sitting at the bottom;
# random_state=1 makes the shuffle reproducible, reset_index renumbers the rows cleanly
co_df_with_dupes = co_df_with_dupes.sample(frac=1, random_state=1).reset_index(drop=True)

# Report how many rows we have now
print("Rows:", len(co_df_with_dupes))

# Count rows that are duplicates based on the Date and Time columns combined;
# .sum() counts the True values in the resulting boolean column
print("Duplicate count:", co_df_with_dupes.duplicated(subset=['Date', 'Time']).sum())
````

Let's now remove these duplicates. 

````python
# Remove duplicate rows, keeping the first occurrence of each Date/Time pair;
# inplace=True modifies the DataFrame directly rather than returning a new one
co_df_with_dupes.drop_duplicates(subset=['Date', 'Time'], keep='first', inplace=True)

# Confirm the duplicates are gone
print("Rows after deduplication:", len(co_df_with_dupes))
````

### Standardising Data Formats

It is common to receive date and time as separate columns. Combining them into a single datetime column simplifies time-series analysis.


````python
# Reload the dataset again
co_df = pd.read_csv('carbonmonoxide-qld-2022.csv', sep=',')
print(co_df)
````

````python
# Convert the Date column from a string (e.g. '01/01/2022') to a proper datetime object
co_df['Date'] = pd.to_datetime(co_df['Date'], format='%d/%m/%Y')

# Convert the Time column from a string (e.g. '14:00') to a timedelta by appending
# the seconds component ':00', allowing it to be added to a date
co_df['Time'] = pd.to_timedelta(co_df['Time'] + ':00')

# Add Date and Time together to produce a single combined timestamp column
co_df['DateTime'] = co_df['Date'] + co_df['Time']

# Remove the now-redundant separate Date and Time columns
co_df.drop(columns=['Date', 'Time'], inplace=True)

# Reorder columns so DateTime appears first; the list comprehension collects all
# other column names, excluding DateTime, to place after it
co_df = co_df[['DateTime'] + [c for c in co_df.columns if c != 'DateTime']]

print(co_df.head())
````

## Descriptive Statistics

Descriptive statistics summarise the key features of a dataset. The two main categories are:

- **Measures of central tendency**: mean, median, mode
- **Measures of dispersion**: standard deviation, variance

### Mean

$$\bar{X} = \frac{1}{N} \sum_{i=1}^{N} x_i$$

The mean is sensitive to outliers. We will work with the Woolloongabba column (column index 3) to manually calculate and then check with NumPy.


````python
fourth_column = co_df.iloc[:, 3]

sum_values    = fourth_column.sum()
number_values = fourth_column.count()   # non-missing values only
mean          = sum_values / number_values

print("Manual mean:", mean)

#...or in one line with Pandas
print("Pandas mean:", fourth_column.mean())
````

### Median

The median is the middle value of a sorted dataset. It is more robust to outliers than the mean.

$$\text{Median} = \begin{cases} X[\tfrac{N+1}{2}] & N \text{ odd} \\ \tfrac{X[N/2] + X[N/2+1]}{2} & N \text{ even} \end{cases}$$


````python
sorted_values = fourth_column.dropna().sort_values().reset_index(drop=True)  # drop NaNs, sort ascending, reset row numbers

N = len(sorted_values)  # number of non-NaN values
print(N)

# N is even, so the median is the average of the two middle values at positions N/2 and N/2+1
median = (sorted_values.iloc[N//2 - 1] + sorted_values.iloc[N//2]) / 2

print("Manual median:", median)

#...or in one line with Pandas
print("Pandas median:", fourth_column.median())
````

### Mode

The mode is the most frequently occurring value.


````python
values_count = Counter(fourth_column.dropna())
max_count = max(values_count.values())
modes = [v for v, c in values_count.items() if c == max_count]

print("Manual mode:", modes)

#...or in one line with Pandas
print("Pandas mode:", fourth_column.mode().values)
````

### Standard Deviation

For a **sample**:

$$s = \sqrt{\frac{1}{N-1} \sum_{i=1}^{N} (x_i - \bar{X})^2}$$

Setting `ddof=1` in pandas/NumPy selects the sample formula (divide by $N-1$); `ddof=0` gives the population formula (divide by $N$).


````python
variance = ((fourth_column - mean) ** 2).sum() / (number_values - 1)
std_dev  = variance ** 0.5

print("Manual std dev:", std_dev)

#...or in one line with Pandas
print("Pandas std dev:", fourth_column.std())
````

Notice how Pandas made everything so much easier to calculate. Another quick approach...

## Inferential Statistics

Inferential statistics let us draw conclusions about populations from samples. Here we use a simple two-sample t-test to ask whether the difference in mean CO levels between Woolloongabba and South Brisbane is likely to reflect a real difference, or could have arisen by chance.

This is intended as a practical introduction to running statistical tests in Python. A t-test carries assumptions, most importantly that the data are approximately normally distributed and that the two groups have similar variances, which our CO data does not strictly satisfy. In practice you would check these assumptions before proceeding, and may need to use alternative tests as a result. For a more complete treatment, refer to your statistics unit or the [SciPy statistics documentation](https://docs.scipy.org/doc/scipy/reference/stats.html), which covers a wide range of tests and when to use them.

````python
from scipy.stats import ttest_ind

woollongabba_clean   = co_df.iloc[:, 3].dropna()  # Woolloongabba CO values, NaNs removed
south_brisbane_clean = co_df.iloc[:, 2].dropna()  # South Brisbane CO values, NaNs removed

# Two-sample t-test; equal_var=False uses Welch's variant, which is safer when
# you are not certain the two groups share the same variance
t_stat, p_value = ttest_ind(woollongabba_clean, south_brisbane_clean, equal_var=False)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value:     {p_value:.3e}")

if p_value < 0.05:
    print("Result: significant difference between the two locations (p < 0.05).")
else:
    print("Result: no significant difference detected (p >= 0.05).")
````

The t-statistic measures how large the difference between the two group means is relative to the variability in the data. The p-value is the probability of observing a difference this large if there were truly no effect, a small p-value (conventionally < 0.05) gives us reason to reject that assumption.

In our case, a t-statistic of -15.8 and a p-value of 8.2e-56 gives very strong evidence that the two locations have different mean CO levels. The negative t-statistic simply reflects the order the groups were passed in, Woolloongabba was subtracted from South Brisbane, and tells us Woolloongabba has the lower mean of the two.


### Confidence Intervals

A confidence interval gives a range of plausible values for the true population mean, rather than just a single point estimate. A 95% confidence interval means that if we repeated the sampling many times, 95% of the resulting intervals would contain the true mean.

$$CI = \bar{X} \pm t_{\alpha/2} \cdot \frac{s}{\sqrt{N}}$$

where $\bar{X}$ is the sample mean, $s$ is the sample standard deviation, $N$ is the sample size, and $t_{\alpha/2}$ is the critical value from the t-distribution for the chosen confidence level.

````python
import scipy.stats as stats
import numpy as np
from scipy.stats import t, sem

# Manual approach using the CI formula
def confidence_interval(data, confidence=0.95):
    n  = len(data)                                     # sample size
    m  = data.mean()                                   # sample mean
    se = data.std() / np.sqrt(n)                       # standard error of the mean
    tc = stats.t.ppf((1 + confidence) / 2, df=n - 1)  # critical t-value
    return m - tc * se, m + tc * se                   # lower and upper bounds

print("Manual:")
for label, grp in [("Woolloongabba", woollongabba_clean), ("South Brisbane", south_brisbane_clean)]:
    lo, hi = confidence_interval(grp)
    print(f"  {label}: ({lo:.4f}, {hi:.4f})")

# ...or in one line using scipy
print("\nSciPy:")
for label, grp in [("Woolloongabba", woollongabba_clean), ("South Brisbane", south_brisbane_clean)]:
    lo, hi = t.interval(0.95, df=len(grp)-1, loc=grp.mean(), scale=sem(grp))
    print(f"  {label}: ({lo:.4f}, {hi:.4f})")
````

Both methods return identical results, confirming our manual implementation is correct.

The confidence intervals tell us not just where the mean sits, but how precisely we have estimated it. The intervals for Woolloongabba (0.1512, 0.1561) and South Brisbane (0.1822, 0.1883) do not overlap, which is a strong visual indicator of a significant difference, consistent with what the t-test reported above.

Note that a narrow confidence interval reflects a large sample size or low variability, not necessarily that the data are well-behaved statistically. As with the t-test, the validity of this interval relies on assumptions about the underlying distribution. For a more robust introduction to statistical inference in Python, the [Statsmodels documentation](https://www.statsmodels.org/stable/index.html) is a good next reference.

## Visualising Uncertainty and Spread

In physics, a plot without uncertainty information is usually incomplete. Three common approaches are summarised below:

| Method | Shows | When to use |
|--------|-------|-------------|
| Error bars | SD, SE, or CI | Discrete group comparisons in bar and scatter plots |
| Shaded bands | CI or SD around a curve | Continuous data such as line plots, model fits, and time series |
| Boxplot / violin | Full distribution shape | Comparing the spread and skew of distributions across groups |

Choosing the right method depends on what you want to communicate:

- **Error bars** are most familiar and work well when you have a small number of discrete groups or data points. They can become cluttered on dense scatter plots.
- **Shaded bands** are natural for continuous data where you have a curve or model prediction and want to show the uncertainty around it across the full range.
- **Boxplots and violin plots** are the best choice when the shape of the distribution matters, not just the mean. They immediately reveal skew, spread, and outliers in a way that a single error bar cannot.

**Important:** always label which measure you are showing (SD vs SE vs CI). They convey different things and are not interchangeable.

- **Standard deviation (SD)**: describes the spread of the data itself or how much individual measurements vary.
- **Standard error (SE = SD / √N)**: describes uncertainty in the estimated mean; shrinks as sample size increases.
- **Confidence interval**: a range within which the true mean lies with a specified probability, accounting for both spread and sample size.

````python
categories = ['A', 'B', 'C']
values     = [5, 7, 3]                      # mean value for each category
std_dev    = [0.5, 0.7, 0.3]                # standard deviation for each category, calculated elsewhere

plt.figure(figsize=(5, 4))
plt.bar(categories, values,
        yerr=std_dev,                       # add vertical error bars of the specified size above and below each bar
        capsize=5,                          # width of the horizontal cap at the end of each error bar
        color='skyblue')
plt.title('Bar Chart with SD Error Bars')
plt.ylabel('Value ± SD')                    # label the axis to make clear the error bars represent SD
plt.show()
````

````python
x     = np.linspace(0, 10, 100)                             # 100 evenly spaced x values from 0 to 10
y     = np.sin(x)                                           # model curve
error = 0.1 + 0.1 * np.sqrt(x)                              # uncertainty that grows with x, as is common in real measurements

plt.figure(figsize=(7, 4))
plt.plot(x, y, color='blue', label='Model')
plt.fill_between(x, y - error, y + error,                   # shade the region one error width above and below the curve
                 color='blue', alpha=0.2, label='±1σ band')
plt.title('Line Chart with Uncertainty Band')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
````
